```python
import os
import openai
from PyPDF2 import PdfReader
import json

openai.api_key = "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"

# Function to extract text from PDF using PyPDF2
def extract_text_from_pdf(pdf_file):
    text = ""
    with open(pdf_file, 'rb') as f:
        pdf_reader = PdfReader(f)
        for page_num in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page_num].extract_text()
    return text

# Folder containing PDF files
pdf_folder = '/content/sample_data/papers'

# Initialize an empty dictionary to hold the responses for each paper
all_responses = {}

# Iterate over each PDF file in the folder
for idx, file_name in enumerate(os.listdir(pdf_folder)):
    if file_name.endswith('.pdf'):
        file_path = os.path.join(pdf_folder, file_name)
        pdf_text = extract_text_from_pdf(file_path)
        questions = [
            "What is the main objective or research question of the paper?",
            "Can you summarize the key literature and background provided?",
            "What methodology is used in the study?",
            "What are the key findings of the research?",
            "How do the authors interpret their findings?",
            "What limitations do the authors acknowledge about their study?",
            "How does this research contribute to the existing knowledge about cancer or tumors?",
            "What future research directions do the authors suggest?",
            "Can you identify any potential ethical concerns in the study?",
            "What are the statistical tools and methods used, and are they appropriate for this study?",
            "Are the references cited in the paper relevant and current? ",
            "Can you identify any conflicting theories or debates highlighted in the paper?",
            "How might the findings of this paper be relevant to clinical practice or policy in cancer care?" ,
            "Are there any inconsistencies or contradictions within the paper?"
        ]
        # Combine all questions with PDF text
        user_msgs = [pdf_text + "\n\n" + question for question in questions]
        system_msgs = ["You are a helpful info retrieval." for _ in questions]

        # Make a request to OpenAI's chat completion API for all questions at once
        responses = openai.ChatCompletion.create(
            model="gpt-4-0125-preview",
            messages=[
                {"role": "system", "content": system_msgs},
                {"role": "user", "content": user_msgs},
            ],
            max_tokens=150  # Adjust max_tokens as needed
        )

        # Extract responses from the API response
        responses_list = [(questions[i], response.choices[i].message.content) for i, response in enumerate(responses.choices)]

        # Assign a paper name like Paper1, Paper2, etc.
        paper_name = f"Paper{idx + 1}"
        all_responses[paper_name] = responses_list

# Save responses to a JSON file
output_file = 'all_responses.json'
with open(output_file, 'w') as json_file:
    json.dump(all_responses, json_file, indent=4)

print(f"All responses saved to {output_file}")
```
